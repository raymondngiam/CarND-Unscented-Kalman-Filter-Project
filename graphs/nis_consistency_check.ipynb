```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
log_nis_radar = pd.read_csv("../build/nis_radar.csv")
```

```python
log_nis_radar.describe()
```

```python
fig = plt.figure(figsize=(9, 6))
plt.xlabel("time step, k")
plt.ylabel("NIS")
plt.grid()
plt.plot(log_nis_radar[1:], '-', color="c",label="NIS Radar")
plt.hlines(7.81,0,250, color="r",label="5% (7.81, df = 3)")
plt.legend(loc=1) 
plt.title("NIS Radar")
plt.show()
```

```python
log_nis_radar_select = log_nis_radar[1:]
outlier_radar=log_nis_radar_select.loc[log_nis_radar_select["nis_radar"]>7.81]
print(outlier_radar)
```

```python
print("Total time steps: %d"%(len(log_nis_radar_select)))
print("Outlier count: %d"%(len(outlier_radar)))
print("Percentage: %.3f"%(float(len(outlier_radar))/len(log_nis_radar_select)))
```

```python
log_predictions = pd.read_csv("../build/predictions.csv")
```

```python
log_predictions.head()
```

```python
def calc_rmse(data1,data2):
    rmse=data1-data2
    rmse=rmse*rmse
    rmse=np.sqrt(rmse.mean())
    return rmse

rmse_x=calc_rmse(log_predictions["x"],log_predictions["x_gt"])
rmse_y=calc_rmse(log_predictions["y"],log_predictions["y_gt"])
rmse_vx=calc_rmse(log_predictions["vx"],log_predictions["vx_gt"])
rmse_vy=calc_rmse(log_predictions["vy"],log_predictions["vy_gt"])

print("rmse_x: %.4f"%(rmse_x))
print("rmse_y: %.4f"%(rmse_y))
print("rmse_vx: %.4f"%(rmse_vx))
print("rmse_vy: %.4f"%(rmse_vy))
```

```python
def plot_state_ground_truth(log,name):
    plt.xlabel("time step, k")
    plt.ylabel(name)
    plt.grid()
    plt.plot(log[name], '-', color="r",label=name)
    plt.plot(log[name+"_gt"], '-', color="b",label=name+" ground truth")
    plt.legend(loc="best") 
    plt.title(name)

fig = plt.figure(figsize=(14, 12))
fig.add_subplot(2,2,1)
plot_state_ground_truth(log_predictions,"x")
fig.add_subplot(2,2,2)
plot_state_ground_truth(log_predictions,"y")
fig.add_subplot(2,2,3)
plot_state_ground_truth(log_predictions,"vx")
fig.add_subplot(2,2,4)
plot_state_ground_truth(log_predictions,"vy")

plt.show()
```
